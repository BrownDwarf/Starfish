# Julia Matrix Evaluation

This is a notebook to mess around with Julia's matrix implementations to see what the fastest way to code the covariance matrix might be.

Note that we need to solve the problem 

$$\chi^2 = [Y - A(X)]^T C^{-1} [Y - A(X)]$$

where $Y$ is the data vector and will remain constant throughout the problem. $X$ is the independent variable, and $A(X)$ is the (non-linear) predicted model. $C$ is the covariance matrix, which is generally sparse and is Hermitian, Tridiagonal, and positive symmetric definite (all good things)!

Sparse matrices are created using ``sparse``.

## Inverse of Cholesky Decomposition as mentioned on "Don't invert that Matrix"

**Question**: What if I need to multiply something by the inverse, instead of just solving linear equations? For example, say I have a positive definite symmetric matrix $\Sigma$. If I need to calculate a multivariate normal density with this matrix as the covariance, I need to do something like ($X^T \cdot \Sigma^{-1} \cdot X$), where $X$ is a column vector. Can I do this w/o explicitly taking the inverse?

**Answer**: Yes, you can do that by taking the inverse of the Cholesky decomposition. I.e., 

1. decompose as $ L^T \cdot L = \Sigma$, 
2. solve $ L \cdot y = x$ for $y$ (which is fast because $L$ is triangular), (test solutions of triangular matrices vs. full matrices)
3. and then compute $y^T \cdot y$. 
4. Or you could invert $L$ (which is easy because it’s triangular).

## Covariance matrix

Sparse arrays are created with the command ``sparse``, which takes in lists of values from which it will initialize the matrix. For our purposes, we will need to first define a function which can identify the indexes that have an element, solve for the value, then initialize this into the array. Because duplicates are added by default, maybe we could simply create the speye matrix, then create the sparse matrix, then add the two together? Or just initialize everything with ``sparse`` and let them add.

We use a more general covariance function, which is the squared exponential with a Gaussian taper

$$k(x, x^\prime | h, a, \mu, \sigma) = \exp \left ( \frac{-( x - x^\prime)^2 }{2 h^2} \right ) \frac{a^2}{2 \pi \sigma} \exp \left ( - \frac{[(x - \mu)^2 + (x^\prime - \mu)^2]}{2 \sigma^2}\right )$$

here $h$ is a "bandwidth" that controls the power of the oscillations. If $h$ is small, then there will be high-frequency structure. If $h$ is large, then only low-frequency structure will remain.

```julia
function k(x0, x1, h, amp, mu, sigma)
    return amp^2/(2 * pi * sigma^2) * exp(-((x0 - mu)^2 + (x1 - mu)^2)/(2 * sigma^2)) * exp(-0.5 * (x0 - x1)^2/h^2)
end
```

```julia
function C(h, amp, mu, sigma, var=1, size=2300)   
    #For the given mu and sigma, find all of the indexes in the array that will be non-zero. A square aronud mup.
    low = iceil(mu - 4sigma)
    high = ifloor(mu + 4sigma)
    len = high - low + 1
    I = Int[i for i = low:high]
    #Empty arrays that will be filled with Indexes and Values
    II = Array(Int, len^2) 
    JJ = Array(Int, len^2) 
    SS = Array(Float64, len^2) 
    ind = 1
    for i = low:high, j = low:high
        II[ind] = i
        JJ[ind] = j
        SS[ind] = k(i, j, h, amp, mu, sigma)
        ind += 1
    end
    return sparse(II, JJ, SS, size, size) + speye(Float64, size)
end
                
```

```julia
@time C(10, 10, 20., 3.3, 1, 100);
```

```julia
#Plot the results
mysparse = C(10, 10, 20., 3.3, 1, 100)
imshow(mysparse, origin="upper")
```

```julia
#Show the 'sparsity' of the matrix
spy(mysparse, markersize=4, marker=".")
```

# Computing $\chi^2$ using the sparse matrix

create the matrix

```julia
n = 2300
S = C(10, 10, 20., 3.3, 1, n);
xs = linspace(5000, 5200, n);
ys = ones((n,));
```

```julia
#Compute the Chi^2 using Cholesky factorization
function chi2()
    cholobj = cholfact(S);
    out = cholobj \ ys;
    dot(ys, out[:,1])
end
```

```julia
@time chi2()
```

```julia
#Compute the Chi^2 with no Cholesky factorization
function chi2nofact()
    out = S \ ys;
    dot(ys, out[:,1])
end
```

```julia
@time chi2nofact()
```

The result is that doing the Cholesky factorization first results in a ~5 times speedup for a 2,300 sized array.
