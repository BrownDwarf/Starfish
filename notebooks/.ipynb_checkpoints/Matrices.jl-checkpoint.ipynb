## Cholesky Decomposition

Press Ch 2.9. If a square matrix $A$ happens to be symmetric and positive-definite (like our covariance matrix) hten we can decompose it into

$$L \cdot L^T = A$$

Which makes the solution of $A \cdot x = B$ that much eaiser.

1. solve $L \cdot y = b$
2. then solve $L^T \cdot x = y$

Both of these might be easier with Julia, for example using ``A//B``. It really seems like we will just have to play around with all of these.

## Inverse of Cholesky Decomposition as mentioned on "Don't invert that Matrix"

**Question**: What if I need to multiply something by the inverse, instead of just solving linear equations? For example, say I have a positive definite symmetric matrix $\Sigma$. If I need to calculate a multivariate normal density with this matrix as the covariance, I need to do something like ($X^T \cdot \Sigma^{-1} \cdot X$), where $X$ is a column vector. Can I do this w/o explicitly taking the inverse?

**Answer**: Yes, you can do that by taking the inverse of the Cholesky decomposition. I.e., 

1. decompose as $ L^T \cdot L = \Sigma$, 
2. solve $ L \cdot y = x$ for $y$ (which is fast because $L$ is triangular), (test solutions of triangular matrices vs. full matrices)
3. and then compute $y^T \cdot y$. 
4. Or you could invert $L$ (which is easy because it’s triangular).

## Covariance matrix

Sparse arrays are created with the command ``sparse``, which takes in lists of values from which it will initialize the matrix. For our purposes, we will need to first define a function which can identify the indexes that have an element, solve for the value, then initialize this into the array. Because duplicates are added by default, maybe we could simply create the speye matrix, then create the sparse matrix, then add the two together? Or just initialize everything with ``sparse`` and let them add.

We use a more general covariance function, which is the squared exponential with a Gaussian taper

$$k(x, x^\prime | h, a, \mu, \sigma) = \exp \left ( \frac{-( x - x^\prime)^2 }{2 h^2} \right ) \frac{a^2}{2 \pi \sigma} \exp \left ( - \frac{[(x - \mu)^2 + (x^\prime - \mu)^2]}{2 \sigma^2}\right )$$

here $h$ is a "bandwidth" that controls the power of the oscillations. If $h$ is small, then there will be high-frequency structure. If $h$ is large, then only low-frequency structure will remain.

# Covariance functions

## Matern kernels

$$ k_{\nu = 3/2}(r, a, l) = a \left(1 + \frac{\sqrt{3} r}{l} \right ) \exp \left (- \frac{\sqrt{3} r}{l} \right ) $$

$$ k_{\nu = 5/2}(r, a, l) = a \left(1 + \frac{\sqrt{5} r}{l} + \frac{5 r^2}{3 l^2} \right ) \exp \left (- \frac{\sqrt{5} r}{l} \right ) $$

We would like to use sparse matrices with compact support. Therefore we want to taper the Matern kernel to ensure it goes smoothly to zero at some point.

The Haan window is 

$$ w(r) = 0.5 + 0.5 \cos \left( \frac{\pi r}{r_0} \right) $$

We can also use the Julia function ``spdiagm(B, d[, m, n])``, which functions nearly the same way as python's ``sparse.diags()``. This is probably the quickest way to initialize a matrix with the Matern kernel.

```julia
function sparse_k_3_2(xs, amp, l)
    N = length(xs)
    offset = 0
    r0 = 6. * l
    diags = {}
    
    while offset < N
        #Pairwise calculate rs
        if offset == 0
            rs = zeros(Float64, N)
        else
            rs = abs(xs[offset + 1:] - xs[1:end-offset])
        end
        k = similar(rs)
        if minimum(rs) >= r0
            break
        end
        k = (0.5 + 0.5cos(pi .* rs/r0))  .* amp .* (1 + sqrt(3) .* rs/l) .* exp(-sqrt(3) .* rs/l)
        k[rs .>= r0] = 0
        push!(diags, k)
        offset += 1
    end
        
    #Mirror the diagonals
    front = diags[end:-1:2]
    diags = vcat(front, diags)
#     print(offset)
    offsets = [i for i=(-offset + 1):(offset-1)]
#     print(offsets)
    return spdiagm(diags, offsets)
#     return diags
end
```

```julia
npoints = 2300
xs = linspace(5000, 5100, npoints) + 0.03 * randn(npoints);
ys = randn(npoints)
S = sparse_k_3_2(xs, 1., 5)
```

The Haan window is 

$$ w(r) = 0.5 + 0.5 \cos \left( \frac{\pi r}{r_0} \right) $$

The self-convolved Gaussian is

$$k(x, x^\prime | h, a, \mu, \sigma) = \exp \left ( \frac{-( x - x^\prime)^2 }{2 h^2} \right ) \frac{a^2}{2 \pi \sigma} \exp \left ( - \frac{[(x - \mu)^2 + (x^\prime - \mu)^2]}{2 \sigma^2}\right )$$

The question is then how do we properly set $r$ and $r_0$ such that this smoothly goes to 0 at the edges, as the matrix becomes sparse? Because in this scenario, $r \ne x - x^\prime$

What if, instead of choosing ``r_tap`` as some radial distance, we choose it as whatever the maximum ``wl`` excursion?

```julia
function k_region(x0, x1, h, amp, mu, sigma)
    rx0 = abs(x0 - mu)
    rx1 = abs(x1 - mu)
    r_tap = rx0 > rx1 ? rx0 : rx1 #choose the greater distance
    #r_tap = sqrt((x0 - mu)^2 + (x1 - mu)^2)
    
    r0 = 4.0 * sigma
    taper = 0.5 + 0.5cos(pi * r_tap/r0)
    println("For $x0, $x1, taper is $taper")
    return taper * amp^2/(2 * pi * sigma^2) * exp(-((x0 - mu)^2 + (x1 - mu)^2)/(2 * sigma^2)) * exp(-0.5 * (x0 - x1)^2/h^2)
end
```

```julia
function sparse_region(h, amp, mu, sigma, size=2300)   
    #For the given mu and sigma, find all of the indexes in the array that will be non-zero. 
    #A square around mu. Make sure taper works correctly
    low = iceil(mu - 4sigma)
    high = ifloor(mu + 4sigma)
    len = high - low + 1
    I = Int[i for i = low:high]
    #Empty arrays that will be filled with Indexes and Values
    II = Array(Int, len^2) 
    JJ = Array(Int, len^2) 
    SS = Array(Float64, len^2) 
    ind = 1
    for i = low:high, j = low:high
        II[ind] = i
        JJ[ind] = j
        SS[ind] = k_region(i, j, h, amp, mu, sigma)
        ind += 1
    end
    return sparse(II, JJ, SS, size, size) + speye(Float64, size)
end
                
```

```julia
function sparse_region_C(wl, h, loga, mu, sigma)
    N = length(wl)
    #determine the wl indices that are within 4 sigma of mu
    r0 = 4.0 * sigma;
    a = 10^loga
    
    #this loop should exit with first_ind being equal to the first row,col occurence of the region
    i = 1
    r = r0 + 1 #simply starts larger than r0
    first_ind = 1
    last_ind = N
    while (r > r0)
        r = abs(wl[i] - mu) #how far away are we from mu?
        first_ind = i
        i += 1
    end
    
    #this loop should exit with last_ind being equal to the last row,col occurence (inclusive)    
    while (r < r0)
        last_ind = i 
        i += 1
        r = abs(wl[i] - mu)
    end
    
    println("Region stretches from indices $first_ind to $last_ind.")

    #how many non-zero elements?
    M = (last_ind - first_ind + 1)^2
    println("There are $M non-zero elements.")

    if (M == 0)
        println("There are 0 non-zero elements. Returning null.")
        return 0;
    end
    
    II = Array(Int, M) 
    JJ = Array(Int, M) 
    SS = Array(Float64, M)
    k=1
    
    #we want a square and symmetric matrix
    #Only fill in the lower entries (and diagonal). 
    for i = first_ind:last_ind
        for j = first_ind:i
            #r_mu = sqrt((wl[i] - mu)^2 + (wl[j] - mu)^2)
            #if (r_mu < r0) #If the distance is below our cutoff, initialize
                #Initialize lower triangle
            II[k] = i
            JJ[k] = j
            SS[k] = k_region(wl[i], wl[j], h, a, mu, sigma)
            k += 1
            if i != j
                #Initialiaze upper triangle
                II[k] = j
                JJ[k] = i
                SS[k] = k_region(wl[j], wl[i], h, a, mu, sigma)
                k += 1
            end
            #end
        end
    end
    return sparse(II[1:k-1], JJ[1:k-1], SS[1:k-1], N, N)
end
```

```julia
#Compute the Chi^2 with no Cholesky factorization
function chi2nofact()
    out = S \ ys;
    dot(ys, out[:,1])
end
```
