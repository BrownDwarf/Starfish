# Is the noise in a spectrum correlated?

We seek to characterize the covariance in our spectrum and understand its origin. In order to do this, we use high signal to noise TRES observations of Vega. 

## Synopsis
The "global covariance structure" is not due to any correlation in the actual physical noise itself. We verify this by measuring the autocorrelation of regions of stellar continuum and show that there is no significant correlation.

However, there is indeed a global covariance structure present in the *residuals* of any spectroscopic fit. This is because the model is never a perfect fit to the data. If the model is a perfect fit to the data (as in the case of stellar continuum), then the residuals *will be uncorrelated*. However, most spectral lines (even the best modeled ones) are either under-predicted or over-predicted and thus result in *correlated residuals*.

This is borne out by the fact that the global covariance for the Kurucz spectra (better models) is lower in amplitude and characteristic length than the PHOENIX spectra.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline
```

```python
wl0, fl0 = np.loadtxt("vega.txt", unpack=True)
wl1, fl1 = np.loadtxt("vega1.txt", unpack=True)
```

We seek to measure the correlation, if any, of the noise present in a stellar continuum. We use two separate observations of Vega, separated by about an hour. Here is what the raw spectrum looks like.

```python
plt.plot(wl0,fl0)
plt.plot(wl1,fl1)
plt.xlabel(r"$\lambda$ [\AA]")
plt.ylabel("raw counts")
plt.title("Vega")
plt.ylim(6e4, 9e4)
plt.xlim(5140, 5230)
```

In order to test whether pixels in the spectrum are correlated, we select a region of the stellar continuum that is free of lines. Because we have two separate observations, we can have some confidence that any "structure" in the continuum is dominated by noise and is not actually weak lines. We select the following region of spectrum and subtract the mean continuum level.

```python
ind = (wl0 > 5177.3) & (wl0 < 5182.9)

wlt = wl0[ind]
N = len(wlt)

flt0 = fl0[ind]
flt1 = fl1[ind]

#normalize
flt0 = flt0 - np.mean(flt0)
flt1 = flt1 - np.mean(flt1)
```

```python
plt.plot(wlt, flt0)
plt.plot(wlt, flt1)
plt.xlabel(r"$\lambda$ [\AA]")
plt.ylabel("normalized counts")
plt.title("Vega")
```

Because there is some broad shape to the spectrum, we fit splines to the data set and subtract out this general continuum trend.

```python
#Fit splines and subtract the mean continuum level for spectrum 1

spl0 = UnivariateSpline(wlt, flt0, s=3e7)
fl_spl0 = spl0(wlt)
fig, ax = plt.subplots(nrows=2, sharex=True)
ax[0].plot(wlt, flt0)
ax[0].plot(wlt, fl_spl0)

flt0n = flt0 - fl_spl0
ax[1].plot(wlt, flt0n)
ax[1].set_ylabel("normalized counts")
ax[-1].set_xlabel(r"$\lambda$ [\AA]")
```

```python
#Fit splines and subtract the mean continuum level for spectrum 2

spl1 = UnivariateSpline(wlt, flt1, s=3e7)
fl_spl1 = spl1(wlt)
fig, ax = plt.subplots(nrows=2, sharex=True)
ax[0].plot(wlt, flt1)
ax[0].plot(wlt, fl_spl1)

flt1n = flt1 - fl_spl1
ax[1].plot(wlt, flt1n)
ax[1].set_ylabel("normalized counts")
ax[-1].set_xlabel(r"$\lambda$ [\AA]")
```

As far as the eye can see, there is no obvious line structure to the continuum in the lower panels, and one could say that this region of the spectrum is dominated by noise. The downward trend with increasing wavelength has been removed by fitting the continuum with a smoothing spline and subtracting it off.

To test whether or not there is correlation, we will calculate an estimate of the autocorrelation 

$$\hat{R}(k) = \frac{1}{(n - k) \sigma^2} \sum_{t = 1}^{n - k} (X_t - \mu) (X_{t+k} - \mu) $$

```python
def estimated_autocorrelation(x):
    n = len(x)
    variance = np.var(x)
    x = x - np.mean(x)
    r = np.correlate(x, x, mode = 'full')[-n:]
    #assert N.allclose(r, N.array([(x[:n-k]*x[-(n-k):]).sum() for k in range(n)]))
    result = r/(variance*(np.arange(n, 0, -1)))
    return result
```

```python
#Autocorrelation of first Vega spectrum
plt.plot(estimated_autocorrelation(flt0n), drawstyle="steps-mid")
plt.title("Vega 1")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

```python
#Autocorrelation of second Vega spectrum
plt.plot(estimated_autocorrelation(flt1n), drawstyle="steps-mid")
plt.title("Vega 2")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

So far, it looks like there isn't any real autocorrelation to the two spectra. The autocorrelation should = 1.0 at 0 offset. If there was some correlation, then we would expect the autocorrelation to be positive for the first few pixel offsets. Interestingly Vega 1 looks like there is some correlation on ~5 pixel scales, but this does not appear in Vega 2. This looks like a coincidence.

To test how autocorrelations can vary in the presence of noise, we try generating fake spectra using uncorrelated and then correlated Gaussian noise.

```python
#Try generating our own uncorrelated noise to check
plt.plot(estimated_autocorrelation(np.random.normal(size=N)), drawstyle="steps-mid")
plt.title("Fake uncorrelated noise 1")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

```python
#Try generating our own uncorrelated noise again to check
plt.plot(estimated_autocorrelation(np.random.normal(size=N)), drawstyle="steps-mid")
plt.title("Fake uncorrelated noise 2")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

By generating fake, uncorrelated Gaussian noise and taking the autocorrelation we see that we can reproduce the correlation structure that was present in Vega 1. This is convincing evidence that general spectroscopic noise, measured at the continuum, is *uncorrelated*.

To check our assertion that noise in the continuum is actually uncorrelated, let's see what the autocorrelation of *correlated* noise actually looks like. We will generate noise with a known correlation length.

```python
#Try generating correlated noise using the Matern kernel

def k_3_2(r, l):
    return (1 + np.sqrt(3) * r/l) * np.exp(-np.sqrt(3) * r/l)

@np.vectorize
def hann(r, r0):
    if np.abs(r) < r0:
        return 0.5 + 0.5 * np.cos(np.pi * r/r0)
    else:
        return 0

def k_3_2_hann(r, l):
    return k_3_2(r, l) * hann(r, 6 * l)

def k_3_2_hann_func(x0i, x1i, x0v=None, x1v=None, amp=None, l=None):
    x0 = x0v[x0i]
    x1 = x1v[x1i]
    r = np.abs(x1 - x0)
    return amp**2 * k_3_2_hann(r, l)
```

```python
inds = np.arange(N)
mat_3_2 = np.fromfunction(k_3_2_hann_func, (N, N), x0v=wlt, x1v=wlt, amp=1, l=0.15, dtype=np.int)
```

This is what the covariance matrix looks like. We use similar parameters to that of the WASP14 spectrum.

```python
plt.imshow(mat_3_2, origin="upper", interpolation="none")
```

```python
#Take two random draws from this matrix

fl_cor0 = np.random.multivariate_normal(np.zeros((N,)), mat_3_2)
fl_cor1 = np.random.multivariate_normal(np.zeros((N,)), mat_3_2)
plt.plot(wlt, fl_cor0)
plt.plot(wlt, fl_cor1)
plt.title("Fake correlated noise 1 and 2")
plt.xlabel("wavelength")
plt.ylabel("raw counts")
```

Now we plot the autocorrelations of these correlated noise spectra.

```python
plt.plot(estimated_autocorrelation(fl_cor0), drawstyle="steps-mid")
plt.title("Fake correlated noise 1")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

```python
plt.plot(estimated_autocorrelation(fl_cor1), drawstyle="steps-mid")
plt.title("Fake correlated noise 2")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

As we expected, we see that there is a substantial correlation to the noise. The first few pixel offsets are definitely correlated.

If this is the case, and true spectroscopic noise *is uncorrelated*, then where does the global covariance structure come from in our fit? Let's test this by examining actual residuals from a fit to WASP14 using the PHOENIX spectra. I have chosen a region of the spectrum that has a few spectral lines but does not have any egregious lines that would be marked by a "bad region."

# Testing the residual structure

```python
from StellarSpectra.spectrum import DataSpectrum
```

```python
fig, ax = plt.subplots(nrows=2, sharex=True)
spec2 = DataSpectrum.open("WASP14/WASP-14_2009-06-15_04h13m57s_cb.spec.flux", orders=[22,23])
fl23, fl24 = spec2.fls

median_fl23 = np.median(fl23)

fl23n = fl23/median_fl23

#Model
wl23 = np.load("../residuals/wlsz23.npy")[0]
f23 = np.load("../residuals/fl23.npy")[0]/median_fl23

ind = (wl23 > 5144) & (wl23 < 5150)
wl23 = wl23[ind]
f23 = f23[ind]
fl23n = fl23n[ind]

r23 = (fl23n - f23)

ax[0].plot(wl23, fl23n, "b", label="data", drawstyle="steps-mid")
ax[0].plot(wl23, f23, "r:", label="model", drawstyle="steps-mid")
ax[0].legend(loc="lower left")
ax[0].set_ylabel("normalized flux")

ax[1].plot(wl23, r23, "k", drawstyle="steps-mid")
ax[1].set_ylabel("residuals")
```

```python
#Now estimate the autocorrelation of the residuals 
plt.plot(estimated_autocorrelation(r23), drawstyle="steps-mid")
plt.title("Residuals")
plt.xlabel("pixel offset")
plt.ylabel("autocorrelation")
```

Boom, the autocorrelation shows that the residual pixels are indeed correlated on the ~few pixel length scale, similar to what we saw with the forcibly correlated noise.

These residuals are correlated due to slight mismatches in the line shapes between model and data spectrum. This explains why the covariance fit for the Kurucz spectra has a *smaller amplitude and length scale* than the PHOENIX spectra. If the spectral models fit better (Kurucz) then there will be less residual structure.
