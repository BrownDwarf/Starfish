```julia
import Winston
```

```julia
xs = linspace(-20, 50);
```

```julia
function logistic(x, alpha, beta, gamma, delta)
    return alpha ./(1 + exp(- beta * (x - gamma))) + delta
end

function lnLogistic(x, alpha, beta, gamma, delta)
    return log(logistic(x, alpha, beta, gamma, delta))
end
```

```julia
ls0 = logistic(xs, -1., 1, 10, 1);
lns0 = lnLogistic(xs, -1., 1, 10, 1);
```

```julia
p = Winston.FramedPlot(
title="Logistic Function",
xlabel="x",
ylabel="logistic")
Winston.add(p, Winston.Curve(xs, ls0, color="blue"))
#Winston.add(p, Winston.Curve(xs, ls1, color="red"))
```

```julia
p = Winston.FramedPlot(
title="Logistic Function",
xlabel="x",
ylabel="lnLogistic")
Winston.add(p, Winston.Curve(xs, lns0, color="blue"))
#Winston.add(p, Winston.Curve(xs, ls1, color="red"))
```
