# Likelihood function

Bayes rule. The key is the likelihood function, which describes the probability of your data given your model. This assumes that your model is a representation of how your data was generated.

**ASK**: Can anyone give me an example of a likelihood function? 

* Bernoulli, for coin flips
* Poisson, for X-ray counts
* Student-t, for outliers
* $\chi^2$, for data with Gaussian errors

What are the assumptions that go into $\chi^2$? Gaussian errors, and that you know your errors correctly. Most problems this should be good enough, but what about the cases when it is not? These are usually the most interesting and oftentimes the regime in which we are actually working (because we are cutting-edge researchers).

## Ways in which $\chi^2$ can go wrong

Correlations in the data set. Talk about *Kepler* lightcurve and transit. If you fit a transit model and neglect the correlations, you'll infer the wrong properties of the planet (ask Ben Montet).

So, it turns out that fitting spectra actually suffer from the same problem, and the residuals look correlated.
